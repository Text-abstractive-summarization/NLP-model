```python
"""
train_original.json

- name : char
- delivery_date : date_type
- documents
    └ id : int
    └ category : char
    └ media_type : char
    └ media_sub_type 
    └ media_name 
    └ size : char
    └ char_count : int
    └ publish_date : date_type
    └ title : char
    └ text 
        └ []
            └ index : int
            └ sentence : char
            └ highlight_indices : array
    └ document_quality_scores 
        └ readable : int
        └ accurate : int
        └ informative : int
        └ trustworthy : int
    └ extractive : array
    └ abstractive : array

"""


# (271093, 30122)
```

```python
import json
import pandas as pd
from tqdm import tqdm

def data_extract(url, idx):

    """
    url : string
    idx : list,  
        ex. [0, 500]
    """


    with open(url, 'r') as f: 
        data = json.load(f)

    origin = []
    abstract = []

    for article in tqdm( data['documents'][idx[0]:idx[1]]):    # documents index
        
        sent = ''                                               # origin text 담을 변수
        for sentence in article['text']:     
            # print (text)   
            for text in sentence:
                sent += text['sentence']
        
        origin.append(sent)
        
        if len(article['abstractive']) != 1:
            print (article['id'])                               # 삭제 해도 됨
            for i in article['abstractive']:
                abstract.append(i)
        else:
            abstract.append(article['abstractive'][0])      # abstract data가 1개 이상일 경우 기사 id 출력 

    df = pd.DataFrame({'document':origin, 'label': abstract})

    return df
```
